### Fill missing values appropriately. 

#### Cabin has 60 to 70 percent of its values missing , so its better to drop it.

#### For this we use drop(column_name , axis=1 , inplace=True)
#### if want to delete multiple columns instead of passing a single column name we can pass list having multiple column name.
#### axis - tell whether we drop column or row. For dropping column set its value to 1 and for dropping row set its value 0.
#### inplace - we assign it a boolean value(True or False). If we assign it True then it will make changes in original dataframe(and we can not store that dataframe into a variable) , and if we assign it with False it will create a copy of original dataframe and apply changes to it and we can store that dataframe inside a variable.

```python
df.drop("Cabin" , axis = 1 , inplace = True)
df.head()
```

#### Fillings missing values of Age column by the mean of Age column 

```python
# taking mean of Age column
mean_age = df['Age'].mean()
mean_age
```

```python
df['Age'].fillna(mean_age , inplace=True)
df.isnull().sum()
```

#### Filling missing values of Embarked column.
#### for this we can use "ffill" or "bfill" method
##### ffill - replaces the NULL values with the value from the previous row (or previous column, if the               axis parameter is set to 'columns' ).
##### bfill - method replaces the NULL values with the values from the next row (or next column, if the               axis parameter is set to 'columns' ).

```python
df['Embarked'].fillna(method="ffill" , inplace=True)
df.isnull().sum()
```

# Convert Categorical Variables
### Convert categorical variables to numerical format using appropriate techniques.

### There are to ways to convert categorical variables to numerical format-
### 1- One Hot Encoding
### 2- Label Encoding

#### One Hot Encoding - One-hot encoding is one of the most common encoding methods in machine         learning. This method spreads the values in a column to multiple flag columns and assigns 0 or 1 to them.

#### Label Encoding - Label Encoding is a popular encoding technique for handling categorical variables. In this technique, each label is assigned a unique integer based on alphabetical ordering.


#### Here we are using One Hot Encoding to achieve this we use get_dummies(dataframe , dtype , drop_first)

#### using 'dtype' attribute we can decide we want categorical values to be encoded in 0/1 or True/False.
#### For 0/1 - dtype='int'
#### For True/False - dtype='bool'
#### by default dtype is int

#### using 'drop_first' attribute we drop first encoded column. To achieve this we set its value to True.
#### We need to drop_first because we can predict the value for dropped column from remaining columns. For example - WE have 3 encoded columns named A, B, C and we dropped A , if the one value of B and C is 0 then its obivious that the value A will 1 corresponding to the value of B and C.

#### So we conclude that for a column having n categorical value , (n-1) columns will be created. For column having 3 categorical value , 2 columns will be created.

```python
dummies = pd.get_dummies(df['Embarked'] , dtype='int' ,drop_first=True)
dummies
```

### Combining dummies and our original dataframe to make single dataframe.

#### For achieving this we use concat(list of dataframe to merge , axis=1) , here axis=1 tells to concat on columns and axis=0 tells to concat on rows.

```python
df = pd.concat([df , dummies] , axis=1)
df.head()
```

# Feature Engineering¶
## Create new features from existing ones, if necessary. Drop unnecessary columns.

### Dropping 'Embarked' column as we already changed its categorical value to integer value using get_dummies

#### For this we use drop(column_name , axis=1 , inplace=True)
#### if want to delete multiple columns instead of passing a single column name we can pass list having multiple column name.
#### axis - tell whether we drop column or row. For dropping column set its value to 1 and for dropping row set its value 0.
#### inplace - we assign it a boolean value(True or False). If we assign it True then it will make changes in original dataframe(and we can not store that dataframe into a variable) , and if we assign it with False it will create a copy of original dataframe and apply changes to it and we can store that dataframe inside a variable.

```python
df.drop('Embarked' , axis=1 , inplace=True)
df.head()
```

#### Here we are encoding 'Pclass' and 'Sex' columns.
#### But 'Pclass' has numeric values . Sometimes numeric values can also act as a categorical values like top 3 position , classes in train seats etc. But first we will convert numeric value to string as get_dummies() work with string values.

```python
#for get dummies to be applied data type must be string , so we are converting Pclass from int to str
df['Pclass'] = df['Pclass'].astype(str)
dummies = pd.get_dummies(df[['Sex' , 'Pclass']] , drop_first=True , dtype='int') 
dummies
```

#### Combining dummies and df dataframes.

```python
df = pd.concat([df , dummies] ,axis=1)
df.head()
```

### Dropping Unecessary columns and the columns on which label encoding has been performed.

```python
df.drop(["PassengerId" , "Name" , "Pclass" , "Sex" , "Ticket" , "Fare"] , axis=1 , inplace=True)
df.head()
```

### Making a function named pre_processor() which will basically perform Feature Engineering.

```python
def pre_process(data):
    data['Age'].fillna(mean_age , inplace=True)
    data['Embarked'].fillna(method="ffill" , inplace=True)
    data['Pclass'] = data['Pclass'].astype(str)
    dummies = pd.get_dummies(data[['Sex' , 'Pclass' , 'Embarked']] , drop_first=True , dtype='int')
    data = pd.concat([data , dummies] , axis=1)
    data.drop(["PassengerId" , "Name" , "Pclass" , "Sex" , "Ticket" , "Fare" , "Embarked" , "Cabin"] , axis=1 , inplace=True)
    return data
```

### Loading train dataset again

```python
df_new = pd.read_csv('/kaggle/input/titanic/train.csv')
df_new.head()
```

### Seperating Survived column from datafrme and creating a seperate dataframe out of it.

```python
y = df_new['Survived']
X = df_new.drop('Survived' , axis=1)
X = pre_process(X)
X.head()
```

```python
pre_process(df_new)
```

```python
len(X)
```

# Splitting Data
### Split the dataset into training and validation sets.

#### Taking 70% of dataset as train set and 30% as valid set.

```python
train_len = int(len(X) * 0.7)
valid_len = len(X) - train_len
print(train_len , valid_len)
```

```python
X_train = X.iloc[0:train_len,:]
X_valid = X.iloc[train_len:,:]
print(len(X_train) , len(X_valid))

y_train = y.iloc[0:train_len]
y_valid = y.iloc[train_len:]
print(len(y_train) , len(y_valid))
```

# Model Selection and Training
### Choose a machine learning model and train it on the training data.

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train , y_train)
```

# Model Evaluation
### Evaluate the model on the validation set using appropriate metrics.
### Display the evaluation results.

```python
# total number 0f correct value / total number of values
model.score(X_train , y_train)
```

```python
model.score(X_valid , y_valid)
```
