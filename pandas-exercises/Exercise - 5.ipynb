## Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA.

```python
r1 = rd
r1 = r1[(r1.title.str.contains('Christmas') == True) & (r1.country == 'USA')]
r1['date'] = pd.to_datetime(r1['date'], errors='coerce')
r1['month_number'] = r1.date.dt.month
r1
```

```python
r1 = r1.month_number.value_counts().sort_index()
r1
```

```python
r1.plot(kind="bar" , title="Number of movies realeased in each month having Christmas in ihier title in USA" , ylabel="Number of movies" , xlabel="Month Number")
```
